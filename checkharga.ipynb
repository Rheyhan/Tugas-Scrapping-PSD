```python
name=list(map(lambda x: x.replace("(", "").replace(")", ""), name))
link=list(map(lambda x: x.replace("+", "%2B").replace(" ", "+"), name))
```

```python
print(link)
```

```python
name[1076]
```

```python
pricelist=[]
web="https://iprice.co.id/search/?term="
start=2037
for i in range(start, len(link)):
    clear_output()
    print(f'Start: {name[i]}  | {i}/{len(link)}')
    url=web+str(link[i])
    header= {'User-Agent': str(UserAgent().random), "Accept-Language": "en-US, en;q=0.5"}
    r=requests.get(url, headers=header)
    time.sleep(1)
    soup=BeautifulSoup(r.content, "html5lib").find("div", attrs={"class" : "mw gP O e2 nV h6 bF p3"})
    try:
        if soup.find("div", attrs={"class": "oq gy er d4"}).text=="Kami tidak menemukan produk yang dicari, periksa ejaan dan filter Anda.":
            pricelist.append("No information")
    except:
        thesoup=soup.find("div", attrs={"id" : "product-list"})
        
        if len(thesoup.findAll("a", attrs={"class" :"zA oY pu rg zC kF kG kI"}))==0:
            for j in thesoup.findAll("div", attrs={"class":"zA l4 pu rg zC le id kI"}):
                productname= j.find("h3", attrs={"class" :"zE zF qB"}).text.lower()[1:]
                if productname.startswith(name[i].lower()):
                    temp=[i for i in j.findAll("div")][1].text[2:].split()
                    if len(temp)==4:
                        price=(float(temp[1].replace(".", ""))+float(temp[3].replace(".", "")))/2
                    else:
                        price=temp[1].replace(".", "")
                    pricelist.append(price)
                    break
        else:
            for j in thesoup.findAll("a", attrs={"class" :"zA oY pu rg zC kF kG kI"}):
                productname= j.find("h3", attrs={"class":"zE hd"}).text.lower()[1:]
                if productname.startswith(name[i].lower()):
                    temp=j.find("div", attrs={"class": "z2"}).text[1:].split()
                    if len(temp)==4:
                        price=(float(temp[1].replace(".", ""))+float(temp[3].replace(".", "")))/2
                    else:
                        price=temp[1].replace(".", "")
                    pricelist.append(price)
                    break
    if (len(pricelist)+start)<(i+1):
        pricelist.append("No information")    
```
